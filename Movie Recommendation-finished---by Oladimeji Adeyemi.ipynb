```python
import pandas as pd
import numpy as np
from ast import literal_eval
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
movies = movies.merge(credits)
movies = movies.merge(movies)
```

```python
points = ['cast', 'crew', 'keywords', 'genres']
for point in points:
    movies[point] = movies[point].apply(literal_eval)
```

```python
movies.head(2)
```

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
        else:
            return np.nan
```

```python
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        if len(names) > 3:
            names = names[:3]
        return names
    return[]
```

```python

movies['director'] = movies['crew'].apply(get_director)

for point in points:
    movies[point] = movies[point].apply(get_list)
```

```python
movies[['title', 'cast', 'director', 'keywords']]
```

```python
def clean_movies(x):
    if isinstance(x, list):
        return[str.lower(i.replace(" ", "")) for i in x]
    else:
        if isinstance(x,str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
points = ['cast', 'crew', 'keywords', 'genres']
for point in points:
    movies[point] = movies[point].apply(clean_movies)
```

```python
def create_string(x):
    return ' '.join(str(x['keywords'])) + ' ' + ' '.join(str(x['cast'])) + ' ' + str(x['director']) + ' ' + ' '.join(str(x['genres']))
```

```python
movies['string'] = movies.apply(create_string, axis=1)
```

```python
movies.head(2)
```

```python
movies[['string']].head(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(movies['string'])
```

```python
count_matrix.shape
```

```python
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
movies = movies.reset_index()
indices = pd.Series(movies.index, index=movies['title'])
```

```python
# Function that takes in movie title as input and outputs most similar movies
def get_recommendations(title, cosine_sim=cosine_sim):
    #Index of the movie that matches the title
    idx = indices[title]

    #Pairwsie similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sorts the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    #Scores of the 10 most similar movies
    sim_scores = sim_scores[1:11]

    #Movie indices
    movie_indices = [i[0] for i in sim_scores]

    #Then returns the top 10 most similar movies
    return movies['title'].iloc[movie_indices]
```
